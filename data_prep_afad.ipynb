```python
import argparse
import os
import numpy as np
import utils
import pandas as pd
import re
from glob import glob
```

```python
def clean_path(row: str):
        #print(row.split('/')[5:])
        return '/'.join(row.split('/')[5:])

age_list=[i for i in range(18,40)]
main_data = pd.DataFrame()

for i in age_list:
    for g in [111,112]:
        data = pd.DataFrame()
        full_path=str(i)+'/'+str(g)+'/*.jpg'
        pattern = '/home/zarana/cnn2ranking/AFAD/'+full_path
        paths = glob(pattern)
        data['full_path'] = paths
        data['full_path']=data['full_path'].map(clean_path)
        data['db_name'] ='AFAD'
        data['age']=i
        if g==111:
            data['gender']=1
        else:
            data['gender']=0
        main_data=main_data.append(data, ignore_index=True)

for i in range(10):        
    main_data=main_data.sample(frac=1).reset_index(drop=True)    

main_data
```

```python
main_data.to_csv('AfAD.csv', columns=['db_name', 'full_path', 'age','gender'], index=False)
```
